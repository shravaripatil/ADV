```
#Task Description : plot the line graph to show the 5 states with highest GDP in 2010 and 2014
import pandas as pd
import matplotlib.pyplot as plt 

pd.set_option('display.max_columns', None)
df = pd.read_csv('https://shravaripatil.github.io/ADV/AA/Energy_Census_and_Economic_Data%20US_2010-2014.csv')
df.head(10)

sorted_GDP2010_df = df.sort_values('GDP2010', ascending =False)
sorted_GDP2010_df = sorted_GDP2010_df.head(5)
sorted_GDP2014_df = df.sort_values('GDP2014', ascending =False)
sorted_GDP2014_df = sorted_GDP2014_df.head(5)

#using the plt.figure() function, set the size of the plotting area by mentioning the width and height for the figsize parameter.
plt.figure(figsize=(19,10))

# set the label for the x axis values as 'States' (Hint - Use xlabel() function of pyplot module and pass the label in string format)
plt.xlabel('States')

# set the label for the x axis values as 'GDP' (Hint - Use ylabel() function of pyplot module and pass the label in string format)
plt.ylabel('GDP')

# Set the x ticks direction to vertical (Hint - Use xticks() function of pyplot module and pass the direction in string format to rotation parameter of this function)
plt.xticks(rotation='vertical')

# Set the title of the plot (Hint - Use title() function of the pyplot module and pass the direction in string format)
plt.title('Top 5 Energy Consumption VS Energy production States')

# Draw the line graph 
# Use plot() function od the pyplot module and pass the following parameters 
    # x values - Assign this parameter 'state' names of US (Hint - Access the 'State' column of the dataframe , the dataframe is stored in the 'sorted_GDP2010_df' variable. )
    # y values - Assign this parameter values of 'GDP2010'  (Hint - Access the values of 'GDP2010' column of the dataframe , the dataframe is stored in the 'sorted_GDP2010_df' variable. )
    # label = Assign this parameter a string which tells what does the line denote
    # linewidth - set the line width
plt.plot(sorted_GDP2010_df['State'], sorted_GDP2010_df['GDP2010'], label ='Consumption', linewidth=3.0)

# Draw the line graph 
# Use plot() function od the pyplot module and pass the following parameters 
    # x values - Assign this parameter 'state' names of US (Hint - Access the 'State' column of the dataframe , the dataframe is stored in the 'sorted_GDP2010_df' variable. )
    # y values - Assign this parameter values of 'GDP2014'  (Hint - Access the values of 'GDP2014' column of the dataframe , the dataframe is stored in the 'sorted_GDP2014_df' variable. )
    # label = Assign this parameter a string which tells what does the line denote
    # linewidth - set the line width
plt.plot(sorted_GDP2014_df['State'], sorted_GDP2014_df['GDP2014'], label = 'Production', linewidth=3.0)

# Display the legend for the plot (Hint - Use legend() function of pyplot module to display the plot)
plt.legend()

# Display the plot (Hint - Use show() function of pyplot module to display the plot)
plt.show()
```
